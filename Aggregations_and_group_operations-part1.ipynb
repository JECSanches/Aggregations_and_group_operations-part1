### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
df2 = sinasc_raw[['ESCMAE', 'PESO', 'SEXO', 'DTNASC']]

# Filtrando pelo aniversário
df2_filtrado = df2[df2['DTNASC'] == '2019-05-14']

# O shape do dataframe
print(df2_filtrado.shape)

# Obtendo o valor médio do peso
peso_medio = df2_filtrado.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().to_frame()
peso_medio
```

```python
# Uma forma um pouco diferente

# Filtrando pelo aniversário
df2_filtrado = df2[df2['DTNASC'] == '2019-05-14']

# Removendo a variável str
df2_filtrado = df2_filtrado.drop(columns='DTNASC')

# Obtendo o valor médio do peso
peso_medio = df2_filtrado.groupby(['ESCMAE', 'SEXO']).mean()
peso_medio
```

```python
# Outra forma de visualizar 

peso_medio.unstack()
```

```python
# Visualizando graficamente

peso_medio.unstack().plot(figsize=[7,3])
plt.tick_params(axis='x', rotation=45)
plt.ylabel('peso')
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
# Definindo o dataframe de interesse
df3 = sinasc_raw[['munResNome', 'QTDFILVIVO', 'IDADEMAE', 'IDADEPAI']]

# Município com maior número de nascimentos 
df3.groupby('munResNome')['QTDFILVIVO'].agg('max').sort_values(ascending=True)
```

```python
# Conhecendo o município com menor número de nascimentos "Vale do Paraíso"

#Filtrando pelo município
df3 = df3[df3['munResNome'] == "Vale do Paraíso"]
# O shape 
print(df3.shape)

# Aplicando o método aggregator() ou agg()
# para obter a média, o máximo e o mínimo das idades de pais e mães
df3.groupby('munResNome')[['IDADEMAE', 'IDADEPAI']].agg(['mean', 'max', 'min'])
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
# Definindo o DF
df4 = sinasc_raw[['munResNome', 'QTDFILVIVO', 'IDADEMAE', 'IDADEPAI']]
df4.groupby('munResNome')['QTDFILVIVO'].agg('max').sort_values(ascending=False)
```

```python
# O município com maior número de nascimentos é "Santa Luzia D'Oeste"

# Filtrando pelo município de interesse
df4 = df4[df4['munResNome'] == "Santa Luzia D'Oeste"]

# Obtendo os valores médios, máximos e mínimos
df4.groupby('munResNome').agg(['mean', 'max', 'min'])
```

Um ponto interessante referente ao segundo exercício é a relação de peso entre os bebês de sexo masculino e feminino. Apesar de em alguns casos haver exceção, majoritariamente os bebês masculinos são mais pesados que os femininos e há uma relação entre esse peso e a escolaridade da mãe. Como pode ser visto abaixo.

```python
df = sinasc_raw[['ESCMAE', 'PESO', 'SEXO', 'DTNASC']]
df.groupby(['ESCMAE', 'SEXO'])['PESO'].mean().to_frame().unstack().plot()
plt.ylabel('peso em gramas')
```

Um outro ponto que muito chama a atenção é a discrepância no número máximo de bebês nascidos no município *Santa Luzia D'Oeste*, como apresentado no exercício 4. Sendo que o número máximo seria de 30 bebês, porém o segundo maior número é 4, o que pode nos levar a crer que há um erro nos dados e que possivelmente o valor de 30 é um erro de digitação, sendo um dado outlier.

Na figura abaixo está presente a relação do número de bebês nascidos na cidade de *Santa Luzia D'Oeste* ao longo de 2019 e, como pode-se notar, há um ponto discrepante em relação aos demais. 

```python
df5 = sinasc_raw[['munResNome', 'QTDFILVIVO', 'DTNASC']]
df5 = df5[df5["munResNome"] == "Santa Luzia D'Oeste"]

df5.groupby('DTNASC')['QTDFILVIVO'].mean().plot()
plt.title("Bebês nascidos em Santa Luzia D'Oeste no ano de 2019")
plt.ylabel('número de bebês')
plt.tick_params(axis='x', rotation=45)
```
